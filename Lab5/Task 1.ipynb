## Try Linear Regression just using numpy (*Without Tensorflow/Pytorch or other torch library*). You can optionally use sklearn (if you want)

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Temperature, rainfall, humidity

input = np.array([[73, 67, 43], [91, 88, 64], [87, 134, 58], [102, 43, 37], [69, 96, 70],
                    [73, 67, 43], [91, 88, 64], [87, 134, 58], [102, 43, 37], [69, 96, 70],
                    [73, 67, 43], [91, 88, 64], [87, 134, 58], [102, 43, 37], [69, 96, 70]], 
                    dtype='float32')
print(input.shape)
```

```python
# apple, oranges

target = np.array([[56, 70], [81, 101], [119, 133], [22, 37], [103, 119],
[56, 70], [81, 101], [119, 133], [22, 37], [103, 119],
[56, 70], [81, 101], [119, 133], [22, 37], [103, 119]], dtype='float32')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(input, target, test_size=0.3, random_state=88)
```

```python
lr = LinearRegression().fit(X_train, Y_train)
print(lr)
```

```python
print(lr.intercept_)
print(lr.coef_)
```

```python
Y_pred = lr.predict(X_test)
print(Y_pred,"\n")
print(Y_test)
```

```python
from sklearn import metrics

print(f'Mean Squared Error: {metrics.mean_squared_error(Y_test, Y_pred)}')
```
